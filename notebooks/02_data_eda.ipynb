```python
import os
import matplotlib.pyplot as plt
import rasterio
import numpy as np
from PIL import Image
```

## See the downloaded Tiff Files

```python
raster_filepath = r'../data/tile_-121.33760000000002_37.942919999999994.tiff'

# Check if the file exists
if not os.path.exists(raster_filepath):
    print(f"File not found: {raster_filepath}")
else:
    print("File exists!")
```

```python
# Open the raster file
with rasterio.open(raster_filepath) as src:
    raster_data = src.read(3)  # Read the first band

    # Display the raster data using matplotlib
    plt.imshow(raster_data, cmap='viridis')
    plt.colorbar()
    plt.title('Raster Data')
    plt.show()
```
